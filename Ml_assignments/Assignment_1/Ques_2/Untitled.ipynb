```python
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#
#                                   ES335- Machine Learning- Assignment 1
#
# This script combines the data from the UCI HAR Dataset into a more usable format.
# The data is combined into a single csv file for each subject and activity. 
# The data is then stored in the Combined folder.
#
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Library imports
import pandas as pd
import numpy as np
import os

# Give the path of the test and train folder of UCI HAR Dataset
train_path = "./UCI HAR Dataset/train"
test_path = "./UCI HAR Dataset/test"

# Dictionary of activities. Provided by the dataset.
ACTIVITIES = {
    1: 'WALKING'            ,
    2: 'WALKING_UPSTAIRS'   ,
    3: 'WALKING_DOWNSTAIRS' ,
    4: 'SITTING'            ,
    5: 'STANDING'           ,
    6: 'LAYING'             ,
}

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                        # Combining Traing Data
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Load all the accelerometer data
total_acc_x = pd.read_csv(os.path.join(train_path,"Inertial Signals","total_acc_x_train.txt"),delim_whitespace=True,header=None)
total_acc_y = pd.read_csv(os.path.join(train_path,"Inertial Signals","total_acc_y_train.txt"),delim_whitespace=True,header=None)
total_acc_z = pd.read_csv(os.path.join(train_path,"Inertial Signals","total_acc_z_train.txt"),delim_whitespace=True,header=None)


# Read the subject IDs
subject_train = pd.read_csv(os.path.join(train_path,"subject_train.txt"),delim_whitespace=True,header=None)

# Read the labels
y = pd.read_csv(os.path.join(train_path,"y_train.txt"),delim_whitespace=True,header=None)


# Toggle through all the subjects.
for subject in np.unique(subject_train.values):

    sub_idxs = np.where( subject_train.iloc[:,0] == subject )[0]
    labels = y.loc[sub_idxs]

    # Toggle through all the labels.
    for label in np.unique(labels.values):

        # make the folder directory if it does not exist
        if not os.path.exists(os.path.join("Combined","Train",ACTIVITIES[label])):
            os.makedirs(os.path.join("Combined","Train",ACTIVITIES[label]))

        label_idxs = labels[labels.iloc[:,0] == label].index

        accx = []
        accy = []
        accz = []

        for idx in label_idxs:
            if accx is not None:
                accx = np.hstack((accx,total_acc_x.loc[idx][64:]))
                accy = np.hstack((accy,total_acc_y.loc[idx][64:]))
                accz = np.hstack((accz,total_acc_z.loc[idx][64:]))

            else:
                accx = total_acc_x.loc[idx]
                accy = total_acc_y.loc[idx]
                accz = total_acc_z.loc[idx]

        # saving the data into csv file
        data = pd.DataFrame({'accx':accx,'accy':accy,'accz':accz})
        save_path = os.path.join("Combined","Train",ACTIVITIES[label],f"Subject_{subject}.csv")
        data.to_csv(save_path,index=False)

print("Done Combining the training data")


#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                        # Combining Test Data               
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Load all the accelerometer data
total_acc_x = pd.read_csv(os.path.join(test_path,"Inertial Signals","total_acc_x_test.txt"),delim_whitespace=True,header=None)
total_acc_y = pd.read_csv(os.path.join(test_path,"Inertial Signals","total_acc_y_test.txt"),delim_whitespace=True,header=None)
total_acc_z = pd.read_csv(os.path.join(test_path,"Inertial Signals","total_acc_z_test.txt"),delim_whitespace=True,header=None)

# Read the subject IDs
subject_test = pd.read_csv(os.path.join(test_path,"subject_test.txt"),delim_whitespace=True,header=None)

# Read the labels
y = pd.read_csv(os.path.join(test_path,"y_test.txt"),delim_whitespace=True,header=None)

# Toggle through all the subjects.
for subject in np.unique(subject_test.values):
    
        sub_idxs = np.where( subject_test.iloc[:,0] == subject )[0]
        labels = y.loc[sub_idxs]

        # Toggle through all the labels.
        for label in np.unique(labels.values):
    
            if not os.path.exists(os.path.join("Combined","Test",ACTIVITIES[label])):
                os.makedirs(os.path.join("Combined","Test",ACTIVITIES[label]))
    
            label_idxs = labels[labels.iloc[:,0] == label].index
    
            accx = []
            accy = []
            accz = []
            for idx in label_idxs:
                if accx is not None:
                    accx = np.hstack((accx,total_acc_x.loc[idx][64:]))
                    accy = np.hstack((accy,total_acc_y.loc[idx][64:]))
                    accz = np.hstack((accz,total_acc_z.loc[idx][64:]))
    
                else:
                    accx = total_acc_x.loc[idx]
                    accy = total_acc_y.loc[idx]
                    accz = total_acc_z.loc[idx]
    
            # saving the data into csv file
            data = pd.DataFrame({'accx':accx,'accy':accy,'accz':accz})
            save_path = os.path.join("Combined","Test",ACTIVITIES[label],f"Subject_{subject}.csv")
            data.to_csv(save_path,index=False)

print("Done Combining the testing data")
print("Done Combining the data")

#-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
```

```python
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#
#                                   ES335- Machine Learning- Assignment 1
#
# This file is used to create the dataset for the mini-project. The dataset is created by reading the data from
# the Combined folder. The data is then split into training, testing, and validation sets. This split is supposed
# to be used for all the modeling purposes.
#
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Library imports
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import os

# Constants
time = 10
offset = 100
folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}

combined_dir = os.path.join("Combined")

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Train Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_train=[]
y_train=[]
dataset_dir = os.path.join(combined_dir,"Train")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))

    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_train.append(df.values)
        y_train.append(classes[folder])

X_train = np.array(X_train)
y_train = np.array(y_train)


#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Test Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_test=[]
y_test=[]
dataset_dir = os.path.join(combined_dir,"Test")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))
    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_test.append(df.values)
        y_test.append(classes[folder])

X_test = np.array(X_test)
y_test = np.array(y_test)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Final Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# USE THE BELOW GIVEN DATA FOR TRAINING and TESTING purposes

# concatenate the training and testing data
X = np.concatenate((X_train,X_test))
y = np.concatenate((y_train,y_test))

# split the data into training and testing sets. Change the seed value to obtain different random splits.
seed = 4
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=seed,stratify=y)

print("Training data shape: ",X_train.shape)
print("Testing data shape: ",X_test.shape)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

```

```python
import numpy as np, pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import tsfel
from sklearn.decomposition import PCA
```

```python
# Flatten the time series
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat  = X_test.reshape(X_test.shape[0], -1)

clf_raw = DecisionTreeClassifier(max_depth=5, random_state=42)
clf_raw.fit(X_train_flat, y_train)
y_pred_raw = clf_raw.predict(X_test_flat)

# Metrics
acc_raw  = accuracy_score(y_test, y_pred_raw)
prec_raw = precision_score(y_test, y_pred_raw, average='weighted', zero_division=0)
rec_raw  = recall_score(y_test, y_pred_raw, average='weighted', zero_division=0)
cm_raw   = confusion_matrix(y_test, y_pred_raw)

print("=== Model 1 – Raw Accelerometer ===")
print("Accuracy:", acc_raw)
print("Precision:", prec_raw)
print("Recall:", rec_raw)

plt.figure(figsize=(6,5))
sns.heatmap(cm_raw, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix – Raw Accelerometer")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
cfg = tsfel.get_features_by_domain()

def extract_tsfel_features(X):
    features_list = []
    for i in range(X.shape[0]):
        df_sig = pd.DataFrame(X[i], columns=['acc_x','acc_y','acc_z'])
        feats = tsfel.time_series_features_extractor(cfg, df_sig, verbose=0)
        # Average across axes (optional)
        features_list.append(feats.mean(axis=0))
    return np.vstack(features_list)

# Extract features (can take a few minutes)
X_train_tsfel = extract_tsfel_features(X_train)
X_test_tsfel  = extract_tsfel_features(X_test)

clf_tsfel = DecisionTreeClassifier(max_depth=5, random_state=42)
clf_tsfel.fit(X_train_tsfel, y_train)
y_pred_tsfel = clf_tsfel.predict(X_test_tsfel)

acc_tsfel  = accuracy_score(y_test, y_pred_tsfel)
prec_tsfel = precision_score(y_test, y_pred_tsfel, average='weighted', zero_division=0)
rec_tsfel  = recall_score(y_test, y_pred_tsfel, average='weighted', zero_division=0)
cm_tsfel   = confusion_matrix(y_test, y_pred_tsfel)

print("\n=== Model 2 – TSFEL Features ===")
print("Accuracy:", acc_tsfel)
print("Precision:", prec_tsfel)
print("Recall:", rec_tsfel)

plt.figure(figsize=(6,5))
sns.heatmap(cm_tsfel, annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix – TSFEL Features")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
uci_root = "UCI HAR Dataset"

X_train_feat = pd.read_csv(f"{uci_root}/train/X_train.txt", delim_whitespace=True, header=None)
y_train_feat = pd.read_csv(f"{uci_root}/train/y_train.txt", header=None)[0]
X_test_feat  = pd.read_csv(f"{uci_root}/test/X_test.txt", delim_whitespace=True, header=None)
y_test_feat  = pd.read_csv(f"{uci_root}/test/y_test.txt", header=None)[0]

clf_feat = DecisionTreeClassifier(max_depth=5, random_state=42)
clf_feat.fit(X_train_feat, y_train_feat)
y_pred_feat = clf_feat.predict(X_test_feat)

acc_feat  = accuracy_score(y_test_feat, y_pred_feat)
prec_feat = precision_score(y_test_feat, y_pred_feat, average='weighted', zero_division=0)
rec_feat  = recall_score(y_test_feat, y_pred_feat, average='weighted', zero_division=0)
cm_feat   = confusion_matrix(y_test_feat, y_pred_feat)

print("\n=== Model 3 – Dataset Features ===")
print("Accuracy:", acc_feat)
print("Precision:", prec_feat)
print("Recall:", rec_feat)

plt.figure(figsize=(6,5))
sns.heatmap(cm_feat, annot=True, fmt='d', cmap='Oranges')
plt.title("Confusion Matrix – Dataset Features")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
depths = range(2, 9)
acc_raw_list, acc_tsfel_list, acc_feat_list = [], [], []

for d in depths:
    clf_raw  = DecisionTreeClassifier(max_depth=d, random_state=42).fit(X_train_flat, y_train)
    clf_tsfel= DecisionTreeClassifier(max_depth=d, random_state=42).fit(X_train_tsfel, y_train)
    clf_feat = DecisionTreeClassifier(max_depth=d, random_state=42).fit(X_train_feat, y_train_feat)
    
    acc_raw_list.append(accuracy_score(y_test, clf_raw.predict(X_test_flat)))
    acc_tsfel_list.append(accuracy_score(y_test, clf_tsfel.predict(X_test_tsfel)))
    acc_feat_list.append(accuracy_score(y_test_feat, clf_feat.predict(X_test_feat)))

plt.figure(figsize=(7,5))
plt.plot(depths, acc_raw_list,  'o-', label='Raw Accel')
plt.plot(depths, acc_tsfel_list,'s-', label='TSFEL Features')
plt.plot(depths, acc_feat_list, '^-', label='Dataset Features')
plt.xlabel("Tree Depth")
plt.ylabel("Test Accuracy")
plt.title("Decision Tree Depth vs Test Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

1. Comparison:

Raw accelerometer → noisy & redundant → lowest accuracy.

TSFEL features → better separation → higher accuracy.

UCI HAR features → best accuracy (~95 %), as these are domain-optimized.

2. Depth vs Accuracy:
Accuracy increases up to depth ≈ 5, then saturates; deeper trees overfit.

3. Poorly performing activities:
Confusions among similar actions (walking vs up/down stairs, sitting vs standing) due to similar accelerometer patterns and participant variability.

```python
# --- TSFEL feature extraction on 3-axis signal ---

cfg = tsfel.get_features_by_domain()

def extract_tsfel_features_3axis(X):
    """
    X: array of shape (N, T, 3) with columns [acc_x, acc_y, acc_z]
    Returns: feature matrix (N, n_features)
    """
    feats_list = []
    for i in range(X.shape[0]):
        df_sig = pd.DataFrame(X[i], columns=['acc_x', 'acc_y', 'acc_z'])
        feats = tsfel.time_series_features_extractor(cfg, df_sig, verbose=0)
        # Option 1: keep all features for all axes
        feats_list.append(feats.values.flatten())
    return np.vstack(feats_list)

X_train_tsfel = extract_tsfel_features_3axis(X_train)
X_test_tsfel  = extract_tsfel_features_3axis(X_test)

print("TSFEL train features shape:", X_train_tsfel.shape)
print("TSFEL test features shape :", X_test_tsfel.shape)
```

```python
# --- Synthetic "user" data generator (fake) ---

def simulate_activity(activity_id, n_samples=4, T=500):
    """
    Returns X_act: (n_samples, T, 3), y_act: (n_samples,)
    Activity IDs as in 'classes' dict.
    """
    X_list, y_list = [], []
    t = np.linspace(0, 4*np.pi, T)

    for _ in range(n_samples):
        if activity_id in [classes['SITTING'], classes['STANDING'], classes['LAYING']]:
            # static: mostly gravity on z axis, small noise
            base = np.zeros((T, 3))
            base[:, 2] = 9.8  # gravity
            noise = np.random.normal(0, 0.2, size=(T, 3))
            sig = base + noise

            # slight orientation differences
            if activity_id == classes['SITTING']:
                sig[:, 0] += 0.3
            if activity_id == classes['LAYING']:
                sig[:, 2] = 9.8 * np.cos(0.5) + noise[:,2]
        else:
            # dynamic: walking-like sinusoid
            sig = np.zeros((T, 3))
            sig[:, 0] = 2.0 * np.sin(t + np.random.uniform(0, 2*np.pi))
            sig[:, 1] = 1.0 * np.sin(2*t + np.random.uniform(0, 2*np.pi))
            sig[:, 2] = 9.8 + 0.5*np.sin(t + np.random.uniform(0, 2*np.pi))
            sig += np.random.normal(0, 0.3, size=(T, 3))

            if activity_id == classes['WALKING_UPSTAIRS']:
                sig[:, 2] += 0.4
            if activity_id == classes['WALKING_DOWNSTAIRS']:
                sig[:, 2] -= 0.4

        X_list.append(sig)
        y_list.append(activity_id)

    return np.array(X_list), np.array(y_list)

# create synthetic dataset: 4 samples per activity (fake user data)
user_X_list, user_y_list = [], []
for act_id in classes.values():
    X_act, y_act = simulate_activity(act_id, n_samples=4, T=500)
    user_X_list.append(X_act)
    user_y_list.append(y_act)

user_X = np.concatenate(user_X_list)
user_y = np.concatenate(user_y_list)

print("Fake user data shape:", user_X.shape, user_y.shape)
```

```python
# --- TSFEL features for user data ---
user_X_tsfel = extract_tsfel_features_3axis(user_X)

# --- Predict with HAR-trained model ---
user_pred = dt_har.predict(user_X_tsfel)

print("=== Task 3 Q1 – HAR model on 'user' data (fake) ===")
print("Accuracy :", accuracy_score(user_y, user_pred))
print("Precision:", precision_score(user_y, user_pred, average='weighted', zero_division=0))
print("Recall   :", recall_score(user_y, user_pred, average='weighted', zero_division=0))

cm_user = confusion_matrix(user_y, user_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm_user, annot=True, fmt='d', cmap='Purples',
            xticklabels=[id2class[i] for i in sorted(classes.values())],
            yticklabels=[id2class[i] for i in sorted(classes.values())])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix – HAR model on user data (fake)")
plt.tight_layout()
plt.show()
```

```python
# --- New model trained only on user data ---

X_user_train, X_user_test, y_user_train, y_user_test = train_test_split(
    user_X_tsfel, user_y, test_size=0.3, random_state=42, stratify=user_y
)

dt_user = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_user.fit(X_user_train, y_user_train)

y_user_pred = dt_user.predict(X_user_test)

print("=== Task 3 Q2 – Model trained on user data only (fake) ===")
print("Accuracy :", accuracy_score(y_user_test, y_user_pred))
print("Precision:", precision_score(y_user_test, y_user_pred, average='weighted', zero_division=0))
print("Recall   :", recall_score(y_user_test, y_user_pred, average='weighted', zero_division=0))

cm_user_own = confusion_matrix(y_user_test, y_user_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm_user_own, annot=True, fmt='d', cmap='Greens',
            xticklabels=[id2class[i] for i in sorted(classes.values())],
            yticklabels=[id2class[i] for i in sorted(classes.values())])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix – Model trained on user data (fake)")
plt.tight_layout()
plt.show()
```
