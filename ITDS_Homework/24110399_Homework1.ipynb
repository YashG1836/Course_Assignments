#### Ques 1)

Suppose you define a clustering objective in the following manner – give a partitioning C = {C1, . . . Ck},
define

$\text{cost}(\mathcal{C}) = \sum_i \frac{1}{|C_i|} \sum_{x,y \in C_i} \lVert x - y \rVert_2^2$

i.e. cost of a cluster is the sum of all pairwise squared distances. Give an algorithm for this.


#### Answer 1)

Given Points X = {x1,x2,x3.....xn}

Given us are k cluster points. We need an algo which minimizes the cost function given.

On solving the above given cost can be converted in the following form:

$
\boxed{
\frac{1}{|C|} \sum_{x,y \in C} \|x - y\|^2
= 2 \sum_{x \in C} \|x - \mu_C\|^2
}
$

Now from the above formula, its visible that to minimize the cost of the orignal function is same as minimizing distance to cluster mean.

what we can do is , choose k random cluster points, then recursively update the cluster point by taking the mean of the closest point to that cluster.

The given cost function is no different from K-means clustering cost function, and hence it follows a similar algorithm.

Hence acc to K-means algo (LLoyd's Algo):
* Initialize k centers.
* Assign each data point to the nearest center.
* Recompute each center as the mean of its assigned points.
* Repeat until convergence.

#### Proof that given cost is equal to K-means cost

Let $C=\{x_1,\dots,x_n\}$ and let the mean of the cluster be

$$
\mu_C=\frac{1}{n}\sum_{i=1}^n x_i.
$$

Using the identity

$$
\|x-y\|^2=\|x\|^2+\|y\|^2-2x^\top y,
$$

we expand the pairwise cost:

$$
\sum_{x,y\in C}\|x-y\|^2
= 2n\sum_{x\in C}\|x\|^2
-2\left\|\sum_{x\in C}x\right\|^2.
$$

Since $\sum_{x\in C}x = n\mu_C$, this becomes

$$
\sum_{x,y\in C}\|x-y\|^2
=2n\left(\sum_{x\in C}\|x\|^2
-n\|\mu_C\|^2\right).
$$

Now recall the standard identity for k-means:

$$
\sum_{x\in C}\|x-\mu_C\|^2
=\sum_{x\in C}\|x\|^2 - n\|\mu_C\|^2.
$$

Substituting this into the previous expression gives

$$
\sum_{x,y\in C}\|x-y\|^2
=2n\sum_{x\in C}\|x-\mu_C\|^2.
$$

Hence, the pairwise squared-distance cost is equal (up to a factor $2n$) to the usual k-means objective.


#### Ques 2) 

For the k-means problem, show that there is at most a factor of four ratio between the optimal value when
we either require all cluster centers to be data points or allow arbitrary points to be centers.

#### Answer 2)

Let $X$ be the set of data points.  

Let $C^* = \{c_1^*, \dots, c_k^*\}$ be the set of optimal centers that can lie anywhere in $\mathbb{R}^d$.  

Let $J(C^*)$ be the optimal k-means cost (sum of squared errors) using these centers.  

Let $Z = \{z_1, \dots, z_k\} \subseteq X$ be a set of centers chosen only from the data points.  

We compare $J(Z)$ with $J(C^*)$.

Consider one optimal cluster $S \subseteq X$ with center $c^*$.  
The point $c^*$ is the mean of all points in $S$.

Since $c^*$ may not be a data point, choose a point $z \in S$ that is closest to $c^*$.  
Such a point exists because $S$ contains finitely many points.

Because $z$ is the closest data point to the mean,

$$
\|z - c^*\|^2 
\le 
\frac{1}{|S|} \sum_{x \in S} \|x - c^*\|^2 .
$$

Now for any point $x \in S$, write

$$
x - z = (x - c^*) + (c^* - z).
$$

Using the inequality $\|a+b\|^2 \le 2\|a\|^2 + 2\|b\|^2$, we get

$$
\|x - z\|^2 
\le 
2\|x - c^*\|^2 + 2\|z - c^*\|^2 .
$$

Summing over all $x \in S$,

$$
\sum_{x \in S} \|x - z\|^2
\le
2\sum_{x \in S} \|x - c^*\|^2
+
2|S|\|z - c^*\|^2 .
$$

Substituting the earlier bound,

$$
\sum_{x \in S} \|x - z\|^2
\le
2\sum_{x \in S} \|x - c^*\|^2
+
2\sum_{x \in S} \|x - c^*\|^2
=
4\sum_{x \in S} \|x - c^*\|^2 .
$$

Thus, replacing the optimal center $c^*$ by a data point $z$ increases the cost of this cluster by at most a factor of 4.

Applying this argument to all $k$ clusters and adding the costs gives

$$
J(Z) \le 4 J(C^*).
$$

Therefore, even when centers are restricted to lie among the data points, the k-means objective increases by at most a factor of 4.


#### Ques 3) 

Create a random variable X for which Markov’s inequality is tight. Give proof for your answer. If it is
tight, then why are we using other inequalities e.g. Chebyshev and Chernoff?

#### Answer 3)

For any non-negative random variable $X$ and any $a>0$,

$$
P(X \ge a) \le \frac{\mathbb{E}[X]}{a}.
$$

Consider a simple random variable that takes only two values:

$$
X =
\begin{cases}
a & \text{with probability } \frac{1}{a}, \\
0 & \text{with probability } 1 - \frac{1}{a},
\end{cases}
\quad \text{where } a \ge 1.
$$

Its expectation is

$$
\mathbb{E}[X] = a\cdot\frac{1}{a} + 0\cdot\left(1-\frac{1}{a}\right) = 1.
$$

Also,

$$
P(X \ge a) = \frac{1}{a}, \qquad 
\frac{\mathbb{E}[X]}{a} = \frac{1}{a}.
$$

Hence

$$
P(X \ge a) = \frac{\mathbb{E}[X]}{a},
$$

so equality holds and Markov’s inequality is tight for this distribution.  
The intuition is that all the probability mass lies either at $0$ or exactly at the threshold $a$, so the bound cannot be improved.

More generally, any two-point distribution that takes value $a$ with probability $p$ and $0$ otherwise satisfies
$P(X \ge a)=p$ and $\mathbb{E}[X]/a=p$, giving equality.

Even though Markov can be tight for special cases, it is usually quite loose.  
Markov uses only the mean $\mathbb{E}[X]$ and ignores how concentrated the values are. If the variance is small, the true tail probability can be much smaller than the Markov bound.

Chebyshev’s inequality improves this by using variance:

$$
P(|X-\mu|\ge t) \le \frac{\mathrm{Var}(X)}{t^2},
$$

which gives stronger guarantees when the distribution is concentrated.

Chernoff bounds go even further by applying Markov’s inequality to $e^{tX}$:

$$
P(X \ge a) \le \frac{\mathbb{E}[e^{tX}]}{e^{ta}},
$$

and optimizing over $t$ leads to exponentially small tail probabilities, especially for sums of independent random variables.

So Markov is the most general but weakest bound, while Chebyshev and Chernoff use more information about the distribution to obtain much tighter results

#### (a) Cluster them first using k-means clustering, k = 10, with kmeans + + initialization (implement the complete Lloyd’s algorithm yourself). Check the Rand-index of the clustering against the true labels. Use the sklearn module for rand-index.


K–means clustering (k = 10) using Lloyd’s algorithm with k–means++ init

We first flatten each MNIST image from shape $(28,28)$ into a 784-dimensional vector, because k-means works on feature vectors.


```python
import numpy as np
from sklearn.metrics import rand_score, adjusted_rand_score

X = images.reshape(images.shape[0], -1).astype(float)
X = X / 255.0   


def kmeans_plus_plus(X, k, seed=0):
    rng = np.random.default_rng(seed)
    n, d = X.shape
    centers = np.empty((k, d))

    idx = rng.integers(n)
    centers[0] = X[idx]

    for i in range(1, k):
        dist_sq = np.min(((X[:, None, :] - centers[None, :i, :])**2).sum(axis=2), axis=1)
        probs = dist_sq / dist_sq.sum()
        idx = rng.choice(n, p=probs)
        centers[i] = X[idx]

    return centers


def lloyd_kmeans(X, k=10, max_iter=100, tol=1e-4, verbose=True):
    n, d = X.shape
    centers = kmeans_plus_plus(X, k)

    prev_cost = None

    for it in range(1, max_iter + 1):

        distances = ((X[:, None, :] - centers[None, :, :])**2).sum(axis=2)
        cluster_ids = np.argmin(distances, axis=1)

        cost = np.sum(np.min(distances, axis=1))

        if verbose:
            if it%10==0:
                print(f"Iteration {it:3d} | cost = {cost:.4f}")

        new_centers = np.zeros_like(centers)
        for j in range(k):
            pts = X[cluster_ids == j]
            if len(pts) > 0:
                new_centers[j] = pts.mean(axis=0)
            else:
                new_centers[j] = centers[j]

        if np.linalg.norm(new_centers - centers) < tol:
            print(f"\nConverged after {it} iterations.")
            centers = new_centers
            break

        centers = new_centers
        prev_cost = cost

    return cluster_ids, centers, cost, it


pred_clusters, centers, final_cost, num_iter = lloyd_kmeans(X, k=10)

ri  = rand_score(labels, pred_clusters)
ari = adjusted_rand_score(labels, pred_clusters)

print("\nK-Means (Lloyd's) with k=10:")
print(f"  Rand Index: {ri:.4f}")
print(f"  Adjusted Rand Index: {ari:.4f}")
print(f"  Final cost: {final_cost:.2f}")
print(f"  Number of iterations: {num_iter}")
```

#### (b) Do the same for k-center clustering, k = 10. Implement the greedy algorithm discussed in class. Report the Rand-index here too.

Unlike k-means, k-center tries to minimize the maximum distance of any point from its assigned center.  
The greedy algorithm (often called farthest-first traversal) works like this:

1. Pick one random point as the first center.
2. Repeatedly choose the point that is farthest from its nearest existing center.
3. After selecting $k$ centers, assign every point to its nearest center.

Below is a clean implementation from scratch, along with Rand-index evaluation.

```python
import numpy as np
from sklearn.metrics import rand_score, adjusted_rand_score

X = images.reshape(images.shape[0], -1).astype(float)
X = X / 255.0

def greedy_kcenter(X, k=10, seed=0, verbose=True):
    rng = np.random.default_rng(seed)
    n, d = X.shape

    first_idx = rng.integers(n)
    centers_idx = [first_idx]

    if verbose:
        print(f"Initial center index: {first_idx}")

    min_dist_sq = ((X - X[first_idx])**2).sum(axis=1)

    for i in range(1, k):
        next_idx = np.argmax(min_dist_sq)
        centers_idx.append(next_idx)

        if verbose:
            print(f"Chosen center {i}: index {next_idx}, max dist = {min_dist_sq[next_idx]:.4f}")

        new_dist_sq = ((X - X[next_idx])**2).sum(axis=1)
        min_dist_sq = np.minimum(min_dist_sq, new_dist_sq)

    centers = X[centers_idx]

    distances = ((X[:, None, :] - centers[None, :, :])**2).sum(axis=2)
    cluster_ids = np.argmin(distances, axis=1)

    max_radius = np.sqrt(np.max(np.min(distances, axis=1)))

    print(f"\nGreedy k-center completed with k={k}")
    print(f"Maximum cluster radius: {max_radius:.4f}")

    return cluster_ids, centers, max_radius


pred_clusters_kc, kc_centers, radius = greedy_kcenter(X, k=10)

ri  = rand_score(labels, pred_clusters_kc)
ari = adjusted_rand_score(labels, pred_clusters_kc)

print("\nK-Center (Greedy) with k=10:")
print(f"  Rand Index: {ri:.4f}")
print(f"  Adjusted Rand Index: {ari:.4f}")
print(f"  Final max radius: {radius:.4f}")
```

In single linkage, we start with every point as its own cluster and repeatedly merge the two clusters whose closest pair of points has the smallest distance.  
We stop merging once only $k=10$ clusters remain. Then we compare the obtained clusters with the true MNIST labels using the Rand-index.

#### (d) Run the same algorithms (k-means and k-center) but on a rank-k approximation of the training data matrix. Note that if A is the training data matrix (images × pixels), then you can just use UkΣk for the clustering, no need to use Vk. Evaluate for k = 2, 5, 10 and report the rand-index values.

Let $A$ be the data matrix of size (images × pixels).  
Instead of clustering in the original 784-dimensional space, we first compute a rank–k approximation using SVD:

$$
A \approx U_k \Sigma_k V_k^\top.
$$

For clustering, we only use the low-dimensional embedding $U_k \Sigma_k$ (no need for $V_k$).  
We then run k-means and greedy k-center on this representation for $k = 2,5,10$ and compute the Rand-index.

```python
import numpy as np
from sklearn.metrics import rand_score

A = images.reshape(images.shape[0], -1).astype(float) / 255.0

U, S, Vt = np.linalg.svd(A, full_matrices=False)

rank_list = [2, 5, 10]

results = []

for rk in rank_list:

    print(f"\n===== Rank-{rk} approximation =====")

    Uk = U[:, :rk]
    Sk = np.diag(S[:rk])
    A_k = Uk @ Sk  

    pred_km, _, _, _ = lloyd_kmeans(A_k, k=10, verbose=False)
    ri_km = rand_score(labels, pred_km)

    pred_kc, _, _ = greedy_kcenter(A_k, k=10, verbose=False)
    ri_kc = rand_score(labels, pred_kc)

    results.append((rk, ri_km, ri_kc))

    print(f"Rank {rk}:")
    print(f"  Rand Index (k-means)  = {ri_km:.4f}")
    print(f"  Rand Index (k-center) = {ri_kc:.4f}")
print("----------------------------------")
print("Rank |  k-means  |  k-center")
for rk, rkm, rkc in results:
    print(f"{rk:>4} |  {rkm:.4f}   |  {rkc:.4f}")
```

Suppose you have a population of 1 million people, out of which at least 1% are coffee drinkers. You want
to get the estimate of this fraction by using sampling. Give the algorithm and the estimate. What kind
of error bounds can you give with probability 99%?

#### Answer 5) 

To estimate the unknown fraction $p$ of coffee drinkers (assume $p \ge 0.01$), we take a simple random sample.  
Choose a sample size $n$ and pick $n$ people independently at random from the population. Since the population size $N = 10^6$ is very large, treating the samples as independent is reasonable.

For each sampled person define an indicator variable

$$
X_i =
\begin{cases}
1 & \text{if the person drinks coffee}\\
0 & \text{otherwise}
\end{cases}
$$

The estimate of the true fraction is just the average of these values:

$$
\hat{p} = \frac{1}{n}\sum_{i=1}^{n} X_i .
$$

In simple terms, $\hat{p}$ is just the proportion of coffee drinkers in the sample. Because each $X_i$ lies between 0 and 1, we can use the Hoeffding inequality to understand the error:

$$
P(|\hat{p}-p|\ge \epsilon) \le 2e^{-2n\epsilon^2}.
$$

If we want 99% confidence, we set the failure probability $\delta = 0.01$, so

$$
2e^{-2n\epsilon^2}=\delta.
$$

Solving for the error level $\epsilon$ gives

$$
\epsilon=\sqrt{\frac{1}{2n}\ln\left(\frac{2}{\delta}\right)}.
$$

Substituting $\delta=0.01$,

$$
\epsilon=\sqrt{\frac{1}{2n}\ln(200)}
\approx \sqrt{\frac{2.649}{n}}.
$$

So with probability at least 99%, the estimate $\hat{p}$ differs from the true fraction $p$ by at most about $\sqrt{2.649/n}$.
